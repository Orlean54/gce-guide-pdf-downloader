```python
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
HEAD_URL = "https://papers.gceguide.com/A%20Levels/Biology%20(9700)/"

desiredpaperno = ['3','5']

no_of_last_years = 5
```

```python

def getpdfnames(URL):
	print('working')
	req = requests.get(URL)
	href_links = []

	if req.status_code == 200:
		parser_obj = BeautifulSoup(req.content, "html.parser")
		tr_tag_lists = parser_obj.find_all('li', {'class', 'file'})
		for tag in tr_tag_lists:
			all_a_tag = tag.find_all("a", {"class": "name"})
			for a_tag in all_a_tag:
				href_links.append(a_tag["href"])

			#href_links.append(a_tag["href"])
	else:
		print ("Bad status code! Check internet connection")

	print(href_links)

	newlist = []
	for a in href_links:
		if a[-6] in desiredpaperno : newlist.append(a)

	href_links = newlist

	print(newlist)

	for href in href_links:
		req = requests.get("%s/%s" % (URL, href), stream= True)
		if req.status_code == 200:
			link = "%s%s" % (URL,href)
			r = requests.get(link, stream = True)
			filename = href
			
			print(f"{filename} downloaded")
			
			with open(filename, 'wb') as f:
            			for chunk in r.iter_content(chunk_size = 1024*1024):
               				if chunk:
                    				f.write(chunk)

```

```python


def get_all_links(URL):
    req = requests.get(URL)
    yearlist = []

    if req.status_code == 200:
        parser_obj = BeautifulSoup(req.content, "html.parser")
        tr_tag_lists = parser_obj.find_all('a', {'class', 'name'})
        
        for tag in tr_tag_lists:
            yearlist.append(tag.text)
    else:
        print("Bad status code! Check internet connection")
    
    filtered_list = [item for item in yearlist if len(item) <= 4]

    for a in filtered_list[-no_of_last_years:]: #FILTERED YEAR NUMBERS, GET FOR EACH YEAR
        URL = HEAD_URL + f'{a}/'
        getpdfnames(URL)

get_all_links(HEAD_URL)
```
